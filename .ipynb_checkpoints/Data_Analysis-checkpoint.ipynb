#### Remove Duplicates

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Date']
```

```python
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Category")
plt.show()
```
